### creating profit column

```python
data['Profit']=data['Revenue']-data['External commissions']
```

# EDA

```python
comp_sales=data.groupby(data['Company Name'])['Revenue'].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=comp_sales, x='Company Name', y='Revenue')
plt.title('Companies Total Revenue')
plt.xticks(rotation=90)
plt.show()
```

```python
product_profit=data.groupby(data['Brand Name'])['Profit'].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=product_profit, x='Brand Name', y='Profit')
plt.title('Brand Total Profit')
plt.xticks(rotation=90)
plt.show()
```

```python
country_sales=data.groupby(data['ship to country'])['Profit'].sum().nlargest(10).reset_index()
sns.barplot(data=country_sales, x='ship to country', y='Profit')
plt.title('Top 10 Destination Country by Profit')
plt.xticks(rotation=90)
plt.show()
```

## exploring the relation between ship-to and sold-to countries

```python
filter=data.groupby(data['Sold-to party Country Full Name'])['ship to country'].nunique()
fram_filter=filter[filter>1].reset_index()
fram_filter
```

```python
month_sales=data.groupby(data['Months'])['Profit'].sum().reset_index()
sns.lineplot(data=month_sales, x='Months', y='Profit')
plt.title('Monthly Sales Trend')
plt.show()
```

Market Share

```python
#creating market share column
comp_sales['Market Share Value']= comp_sales['Revenue']/comp_sales['Revenue'].sum()
#filtering market share more than 1% for better display
market_share_value=comp_sales.query('`Market Share Value` >0.01')
```

```python
plt.figure(figsize=(10,7))
plt.pie(market_share_value['Market Share Value'], labels=market_share_value['Company Name'], autopct='%1.1f%%')
plt.title('Market Share Value Distribution')
plt.show()
```

```python
#Brand market Share
product_revenue=data.groupby('Brand Name')['Revenue'].sum().sort_values(ascending=False).reset_index()
product_revenue['Market Share']=product_revenue['Revenue']/product_revenue['Revenue'].sum()
#filtering market share more than 1% for better display
product_MS=product_revenue.loc[product_revenue['Market Share']>0.01]
```

```python
plt.figure(figsize=(10,7))
plt.pie(product_MS['Market Share'],labels=product_MS['Brand Name'],autopct='%1.1f%%')
plt.title('Brand Value Market Share Distribution')
plt.show()
```
